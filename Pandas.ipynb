```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style
style.use("fivethirtyeight")


df1=pd.DataFrame({'int_rate':[100,200,300,400],'IND_GDP':[23,12,7,18],"Pop_rate":[1000,2000,3000,4000]},index=[2001,2002,2003,2004])

df2=pd.DataFrame({'int_rate':[110,220,340,470],'IND_GDP':[12,11,17,19],"Pop_rate":[4000,2000,3000,5000]},index=[2003,2004,2005,2006])

#joined=df2.join(df1)

#print(joined)
Concat=pd.concat([df1,df2])
print(Concat)
Concat.plot()
plt.show()

```
